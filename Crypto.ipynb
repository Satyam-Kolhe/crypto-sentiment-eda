```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('historical_data.csv')
df2=pd.read_csv('fear_greed_index.csv')
```

```python
df1
```

```python
df1.info()
df2.info()
```

```python
df2.isnull().sum()
```

```python
df1['Timestamp IST'] = pd.to_datetime(df1['Timestamp IST'], dayfirst=True, errors='coerce')
df1['date'] = df1['Timestamp IST'].dt.date
df2['date'] = pd.to_datetime(df2['date'], dayfirst=True, errors='coerce').dt.date
```

```python
merged = pd.merge(df1, df2, on="date", how="inner")
```

```python
merged
```

```python
# Sentiment distribution
print("Sentiment distribution (number of days):")
print(df2['classification'].value_counts())
```

```python
# Number of trades under each sentiment
print("Number of trades under each sentiment:")
print(merged['classification'].value_counts())
```

```python
#Average and total PnL under Fear vs Greed
print("\PnL summary by sentiment:")
print(merged.groupby("classification")['Closed PnL'].agg(['count','mean','median','sum']))
```

```python
# Average PnL by Side (Long/Short) and sentiment
print("Avg PnL by trade side and sentiment:")
print(merged.groupby(['classification','Side'])['Closed PnL'].agg(['count','mean','median','sum']))
```

```python
#Trade size behavior (USD)
print("Trade Size USD by sentiment:")
print(merged.groupby("classification")['Size USD'].agg(['mean','median','max','min']))
```

```python
# Top and bottom traders
print("Top 5 traders by total PnL:")
print(merged.groupby("Account")['Closed PnL'].sum().sort_values(ascending=False).head(5))

print("Bottom 5 traders by total PnL:")
print(merged.groupby("Account")['Closed PnL'].sum().sort_values().head(5))
```

```python
# Most traded coins under Fear vs Greed
print("Most traded coins by sentiment:")
print(merged.groupby(['classification','Coin']).size().unstack(fill_value=0).head(10))
```

```python
# Fees summary
print("Fee statistics by sentiment:")
print(merged.groupby("classification")['Fee'].agg(['mean','sum','max']))
```
